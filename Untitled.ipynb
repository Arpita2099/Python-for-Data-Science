```python
# Scatter Plot 2D and 3D---- clusters and correlations.
import matplotlib.pyplot as plt
import numpy as np
temperature= np.random.uniform(10,35, size=100)
ice_cream= np.random.uniform(1,50, size=100)
plt.scatter(temperature, ice_cream)
plt.xlabel("Temperature in Degree")
plt.ylabel("Number of ice creams sold")
plt.show()
```

```python
# Scatter Plot 2D and 3D---- clusters and correlations.
import matplotlib.pyplot as plt
import numpy as np
temperature= [10,10,11,12,11,12,13,14,14,13,14.5,15,17,17,18.2,18.8,20,21,21,8,21,22,23]
ice_cream= [1,1,2,2,2,2,1,3,3,2,2,3,4,4,4,4,5,5,5,6,7,8,8]
heater=np.flip(ice_cream)
plt.scatter(temperature, ice_cream, label="Ice Cream")
plt.scatter(temperature, heater, label="Heater")
plt.xlabel("Temperature in Degree")
plt.ylabel("Number of ice creams sold")
plt.legend(loc="upper right")
plt.show()
```

```python
# Histograms---> Universal analysis---> to identify distribution
temperature= [10,10,11,12,11,12,13,14,14,13,14.5,15,17,17,18.2,18.8,20,21,21,8,21,22,23]
plt.hist(temperature, edgecolor='Black')
plt.ylabel("Frequency or Count")
plt.xlabel("Temperatures")
plt.title("Temperature Distribution")
plt.show()
```

```python
data=np.random.normal(loc=80, scale=6, size=1000)
plt.hist(data, edgecolor="black")
plt.show()
```

```python
data=np.random.normal(loc=80, scale=6, size=10000)
plt.hist(data, edgecolor="black", bins=30)
plt.show()
```

```python
print(data) # 68% of the data lies between 64 and 92 with mean -80 and sd=6
count=0
for i in data:
    if 68<=i<=92:
        count+=1
prob= count/10000
print("Total values between 64 and 92 is: ",count)
print("The prob of finding data between 64 and 92 is: ", prob*100, "%")
```

```python
# Boxplot----> quartiles outliers
data=[75,70,72,70,65,80,78,76,60,65,20,100]
plt.boxplot(data)
plt.grid(axis='y')
plt.show()
```

```python
plt.boxplot(data, vert=False)
plt.grid(axis='x')
plt.show()
```

```python
plt.boxplot(data, showmeans= True, patch_artist= True)
plt.grid(axis= 'y')
plt.title("Bloxplot data")
plt.show()
```

```python
my_boxprops={"facecolor":"aqua",
             'color':'pink',
             'linewidth':1}
my_median={ 'color':'teal',
    'linewidth':1}
my_whisker={
    'linestyle':'dashed',
    'alpha':0.7
}
plt.boxplot(data, showmeans= True, patch_artist= True, boxprops=my_boxprops,medianprops= my_median, whiskerprops=my_whisker)
plt.grid(axis= 'y')
plt.title("Bloxplot data")
plt.show()
```
