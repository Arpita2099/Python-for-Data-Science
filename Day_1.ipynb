```python
import pandas as pd
import matplotlib.pyplot as plt
# Sample Data
data={'Month':['Jan','Feb','Mar','Apr'],'Sales':[200, 220, 250, 280]}
df=pd.DataFrame(data)
# Line Plot
df.plot(x='Month',y='Sales',kind='line', marker='o', title='Monthly Sales')
plt.show()
```

```python
# Bar Plots Example
# Sample Data
data={'Category':['A','B','C'],'Values':[10,20,15]}
df=pd.DataFrame(data)
# Bar Chart
df.plot(x='Category', y='Values', kind='bar', title='Category Comparison', color='Blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]
plt.plot(days, steps_walked, "o-g")
plt.plot(days, steps_last_week, "v--m")
plt.title("Step count | This week and last week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
plt.grid(True)
plt.legend(["This week", "Last week"])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the table
days = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]
enfield = [50, 40, 70, 80, 20]
honda = [80, 20, 20, 50, 60]
yahama = [70, 20, 60, 40, 60]
ktm = [40, 20, 30, 70, 60]

# Plotting each bike's data
plt.plot(days, enfield, marker='o', label='Enfield')
plt.plot(days, honda, marker='s', label='Honda')
plt.plot(days, yahama, marker='^', label='Yahama')
plt.plot(days, ktm, marker='d', label='KTM')

# Adding title and labels
plt.title("Bike details in Plot")
plt.xlabel("Days")
plt.ylabel("Distance Covered (in KMs)")

# Adding grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
