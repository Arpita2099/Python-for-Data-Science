```python
def count_genders(count):     #Module 1
    male_count = 0 
    female_count = 0
    for row in count:
        gender = row[5].strip().lower()
        if gender == 'male':
            male_count += 1
        elif gender == 'female':
            female_count += 1
    return male_count, female_count
```

```python
def calculate_average_cgpa(data): #Module 2
    total = 0
    count = 0
    for row in data:
            total += float(row[3])
            count += 1
    return total / count
```

```python
def total_scholarship(data):
    total = 0              #Module 3
    for row in data:
            total += float(row[4])
    return total
```

```python
def read_file(filename):    #module 4
    with open(filename, 'r') as file:
        next(file)
        data = []
        for line in file: 
            temp = [i.strip() for i in line.split(",")]
            data.append(temp)
        return data
```

```python
filename = 'Scholarship.txt'
data = read_file(filename)
print("Males and females are", count_genders(data))
print("Average CGPA is", calculate_average_cgpa(data))
print("Total scholarship amount is", total_scholarship(data))
```
